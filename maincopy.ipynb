```python
X_train, X_test = train[xcols].values, test[xcols].values
y_train, y_test = train[ycols].values, test[ycols].values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler_X = MinMaxScaler().fit(X_train)
Scaler_Y = MinMaxScaler().fit(y_train)
```

```python
X_train_scaled = Scaler_X.transform(X_train)
X_test_scaled = Scaler_X.transform(X_test)

y_train_scaled = Scaler_Y.transform(y_train)
y_test_scaled = Scaler_Y.transform(y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Input

model = Sequential()

model.add(Input(shape=(X_train_scaled.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='linear'))

model.compile(
    loss='mean_squared_error', 
    optimizer='adam', 
    metrics=['mae', 'mse', 'accuracy', 'r2_score'],
    jit_compile='auto'
)

model.summary()
```

```python
EPOCHS = 100

history = model.fit(
    X_train_scaled, y_train_scaled,
    validation_split=0.15,
    epochs=EPOCHS
)
```

```python
model.get_metrics_result()
```

```python
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

mae = history.history['mae']
val_mae = history.history['val_mae']

mse = history.history['mse']
val_mse = history.history['val_mse']
```

```python
plt.figure(figsize=(8, 6))
plt.plot(accuracy, label='Training Accuracy', color='blue', linewidth=7,  alpha=1)
plt.plot(val_accuracy, label='Validation Accuracy', color='orange', linewidth=6,  alpha=1)
plt.xlabel('Epochs',fontsize = 20)
plt.ylabel('Accuracy',fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
#plt.title('Training and Validation Accuracy')
plt.legend(fontsize = 18)
plt.grid(True, linestyle='--', alpha=1)
plt.savefig('./figures/accuracy_plot.png', dpi=300)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, explained_variance_score
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 16})
from sklearn.metrics import PredictionErrorDisplay

def print_metrics(fname, bar, model, x_test_scaled, y_test):
    # Predict the values
    predicted = model.predict(x_test_scaled)

    # Print the metrics section header
    print(f'\n{bar}\n' + '-' * len(bar))
    
    # Display prediction error plot
    display = PredictionErrorDisplay(y_true=y_test, y_pred=predicted)
    
    # Fixed dimensions for the plot
    fixed_width = 8
    fixed_height = 6

    # Plot the prediction error with fixed dimensions
    fig, ax = plt.subplots(figsize=(fixed_width, fixed_height))
    display.plot(kind="actual_vs_predicted", ax=ax)
    
    # Enhancing the plot for IEEE style
    ax.set_title(bar, fontsize=16)  # No title, as IEEE papers often integrate titles within the caption
    ax.set_xlabel("Actual Values", fontsize=16)
    ax.set_ylabel("Predicted Values", fontsize=16)
    
    # Customizing the plot appearance to be clean and professional
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c="gray")  # Adding a diagonal line in gray for subtlety
    ax.legend(['Perfect Prediction', 'Model Prediction'], loc='upper left', fontsize=16, frameon=False)

    # Apply a simple color scheme, IEEE papers typically avoid vivid colors
    ax.lines[0].set_color('black')  # The perfect prediction line in black
    ax.collections[0].set_edgecolor('black')  # Prediction points with black edges
    ax.collections[0].set_facecolor('blue')  # Prediction points in a muted blue

    plt.tight_layout()
    plt.savefig(f'./figures/{fname}.png', dpi=400)
    plt.show()
    
    # Print the metrics
    print(f"R² Score: {r2_score(y_test, predicted):.4f}")
    print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, predicted):.4f}")
    print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, predicted):.4f}")
    print(f"Explained Variance Score: {explained_variance_score(y_test, predicted):.4f}")
    print()
```

```python
from sklearn.dummy import DummyRegressor

dummy_reg = DummyRegressor(strategy="mean")
dummy_reg.fit(X_train_scaled, y_train_scaled)

print_metrics("Ridge","Ridge Linear", dummy_reg, x_test_scaled = X_test_scaled, y_test = y_test_scaled)
```
